```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import openCV
import cv2


img = cv2.imread(r'C:\Users\Data\computer vision\00-puppy.jpg')
type(img)
```

```python
plt.imshow(img)
```

```python
# change image to RGB channel

fix_img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(fix_img);
```

```python
# for grayscale

img_gray = cv2.imread(r'C:\Users\Data\computer vision\00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray);
```

```python
plt.imshow(img_gray, cmap='gray');
```

### Resize Image

```python
fix_img.shape
```

```python
new_img = cv2.resize(fix_img, (1000, 700))
plt.imshow(new_img);
```

```python
new_img.shape
```

```python
# risize by ratio

w_ratio = 0.5
h_ratio = 0.5

new_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio )
plt.imshow(new_img);
```

```python
new_img.shape
```

### Flip Image

```python
flip_img = cv2.flip(fix_img, 0)
plt.imshow(flip_img); # horizotal
```

```python
flip_img = cv2.flip(fix_img, 1)
plt.imshow(flip_img); #vetical
```

```python
flip_img = cv2.flip(fix_img, -1)
plt.imshow(flip_img); # vertical and horizontal both
```

### Save image

```python
cv2.imwrite('Puppy_fix_img.jpg', fix_img)
```
