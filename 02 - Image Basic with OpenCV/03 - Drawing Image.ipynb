```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# draw blank image

blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img);
```

## Draw Rectangle

```python
cv2.rectangle(blank_img, pt1 = (20, 10), pt2=(190,220), color=(255,255,255), thickness= 3)
plt.imshow(blank_img);


# pt1 = top left corner
# pt2 = bottom right corner
```

```python
cv2.rectangle(blank_img, pt1 = (380, 390), pt2=(475, 490), color=(255,0,255), thickness= 3)
plt.imshow(blank_img);
```

```python
cv2.rectangle(blank_img, pt1= (200, 200), pt2=(300,300), color=(255,255,0), thickness=3)
plt.imshow(blank_img);
```

```python
cv2.circle(blank_img, center=(450, 40), radius= 50, color=(0,255,0), thickness=5)
plt.imshow(blank_img);
```

```python
cv2.circle(blank_img, center=(100, 100), radius= 60, color=(255,0,0), thickness= -1)
plt.imshow(blank_img);
```

```python
cv2.circle(blank_img, center=(100, 100), radius= 60, color=(0,0,255), thickness= 4)
plt.imshow(blank_img);
```

### Draw Line

```python
cv2.line(blank_img, pt1 = (0,0), pt2= (512, 512), color=(20, 255,90), thickness= 5)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, pt1=(0,200), pt2=(512, 200), color=(20, 255,255), thickness= 2)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, pt1=(0,350), pt2=(512,350), color=(30, 255,255), thickness= 3)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img ,pt1=(100,512), pt2=(100,0), color=(255,0,255), thickness= 3)
plt.imshow(blank_img)
```

### Write Text

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text='Hello', org=(10, 500), fontFace=font, fontScale= 3, 
            color= (255,255,255), thickness= 5 , lineType=cv2.LINE_AA)

plt.imshow(blank_img)
```

```python
cv2.putText(blank_img, 'Hello', (400, 185), cv2.FONT_HERSHEY_SIMPLEX, 1, (10,0,255),5)
plt.imshow(blank_img)
```

```python
## add polygone

blank_img2 = np.zeros(shape=(512, 512, 3), dtype=np.int16)
plt.imshow(blank_img2);
```

```python
# define vertices of polygon by X-y coordinate

vertices = np.array([[100, 300],  [300,200], [500,300], [500, 300]], dtype=np.int32)

pts = vertices.reshape(-1,1,2)
```

```python
cv2.polylines(blank_img2, [pts],isClosed=True, color=(255,0,0), thickness= 3)
plt.imshow(blank_img2)
```
