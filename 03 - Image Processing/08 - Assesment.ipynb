```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_img(img, cmap=None):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

## TASK: Open and display the giraffes.jpg image that is located in the DATA folder.

```python
giraff = cv2.imread(r'C:\Users\Data\computer vision\giraffes.jpg')
giraff_show = cv2.cvtColor(giraff, cv2.COLOR_BGR2RGB)
display_img(giraff_show)
```

## TASK:Apply a binary threshold onto the image.

```python
giraff_gray = cv2.imread(r'C:\Users\Data\computer vision\giraffes.jpg', 0)

ret1, thresh = cv2.threshold(giraff_gray, 127, 255, cv2.THRESH_BINARY)
display_img(thresh, cmap='gray')
```

## TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)

```python
kernel = np.ones(shape=(4,4), dtype=np.float32)/10
```

```python
kernel
```

```python
giraff = cv2.imread(r'C:\Users\Data\computer vision\giraffes.jpg')
giraff_show = cv2.cvtColor(giraff, cv2.COLOR_BGR2RGB)
```

```python
result = cv2.filter2D(giraff_show, -1, kernel)
display_img(result)
```

## TASK: Create a Horizontal Sobel Filter (sobelx from our lecture) with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.

```python
giraff_gray = cv2.imread(r'C:\Users\Data\computer vision\giraffes.jpg', 0)
```

```python
sobel_x = cv2.Sobel(giraff_gray,cv2.CV_64F, 1,0, ksize=5)
```

```python
display_img(sobel_x , cmap='gray')
```

## TASK: Plot the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image. Pay careful attention to the ordering of the channels.

```python
img = cv2.imread(r'C:\Users\Data\computer vision\giraffes.jpg')

color = ['b','g','r']
for i, col in enumerate(color):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)
```
