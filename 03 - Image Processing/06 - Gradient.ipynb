## Gradient

- An image gradient is a directional change in the intensity or color in an image
-In this section, we will discuss Sobel Feldman Operators and later we will discuss general edge detection

Gradients can be calculated in a specific direction

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12 ,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python

img = cv2.imread(r'C:\Users\Data\computer vision\sudoku.jpg', 0) # grayscale
display_img(img)
```

```python
help(cv2.Sobel)
```

## Sobel X gradient

```python
sobel_x = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=1, 
                    dy = 0, ksize= 5)

display_img(sobel_x)
```

### Sobel Y Gradient

```python
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0,1, 5)
display_img(sobel_y)
```

## laplacian

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)
display_img(laplacian)
```

```python
blanded = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0)
display_img(blanded)
```

```python
#thresgold

ter1, threst = cv2.threshold(img, 100, 200, cv2.THRESH_BINARY)

display_img(threst)
```

```python
# morphology operations

kernel = np.ones((4,4), np.uint8)
mo = cv2.morphologyEx(blanded, cv2.MORPH_GRADIENT, kernel)

display_img(mo)
```
