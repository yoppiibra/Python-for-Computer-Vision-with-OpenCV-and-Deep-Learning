##  Color Mapping

- so far we worked with RGB color spaces
- there are some other models like HSL(Hue,Saturation, Lightness) and HSV (Hue, Saturation, Value)


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r'C:\Users\Data\computer vision\00-puppy.jpg')
plt.imshow(img)  # color BGR
```

```python
# Coverting to RGB
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# converting to HSV
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2HSV))
```

```python
# converting to HSL(HLS)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2HLS))
```
