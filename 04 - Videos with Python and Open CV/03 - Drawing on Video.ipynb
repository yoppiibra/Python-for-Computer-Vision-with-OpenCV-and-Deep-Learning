```python
import cv2


cap = cv2.VideoCapture(0)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))


# for top left corner
x = width // 2
y = height //2


# for the rectangle we want to draw
w = width // 4
h = height // 4

while True:
    ret, frame = cap.read()
    
    cv2.rectangle(frame, (x,y), (x+w, y+h), (255 ,0, 10), 5)
    
    cv2.imshow('frame',frame)
    if cv2.waitKey(10) & 0xff==27:
        break
        
cap.release()
cv2.destroyAllWindows()
```

## Draw Rectangle

```python
import cv2

# callback function
def draw_rec(event, x,y, flags, param):
    global pt1, pt2, topleft_clicked , bottomleft_clicked
    
    if event == cv2.EVENT_LBUTTONDOWN:
        
        if topleft_clicked == True and bottomleft_clicked == True:
            pt1 = (0,0)
            pt2 = (0,0)
            topleft_clicked = False
            bottomleft_clicked = False
            
        if topleft_clicked == False:
            pt1 = (x,y)
            topleft_clicked =True
        
        elif bottomleft_clicked == False:
            pt2 =(x,y)
            bottomleft_clicked = True
            
# global variabel

pt1 = (0,0)
pt2 = (0,0)
topleft_clicked = False
bottomleft_clicked = False

# connect to callback
cap= cv2.VideoCapture(0)
cv2.namedWindow('t')
cv2.setMouseCallback('t', draw_rec)

while True:
    ret, frame = cap.read()
    
    
    # drawing on frame
    if topleft_clicked:
        cv2.circle(frame, center = pt1, radius = 5, color=(0,0,255), thickness =-1)
        
    if topleft_clicked and bottomleft_clicked:
        cv2.rectangle(frame, pt1, pt2, (0,255,0), 4)
        
    cv2.imshow('t', frame)    
    if cv2.waitKey(1) & 0xff == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
    
```
