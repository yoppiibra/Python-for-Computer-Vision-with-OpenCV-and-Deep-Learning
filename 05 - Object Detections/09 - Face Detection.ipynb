## HaarCascade file

```python
face_cascade = cv2.CascadeClassifier(r'C:\Users\Data\computer vision\haarcascades\haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    face_img = img.copy()
    face_rectangle = face_cascade.detectMultiScale(face_img)
    
    for (x,y,w,h) in face_rectangle:
        cv2.rectangle(face_img, (x,y),(x+w, y+h), (255,0,0), 5)
    return face_img

```

```python
result = detect_face(nadia)
```

```python
plt.imshow(result, cmap='gray')
```

```python
result = detect_face(denis)
```

```python
plt.imshow(result, cmap='gray')
```

```python
result = detect_face(solvay)
```

```python
plt.imshow(result, cmap='gray')
```

```python
def detect_face(img):
    face_img = img.copy()
    face_rectangle = face_cascade.detectMultiScale(face_img, scaleFactor = 1.2, minNeighbors= 3)
    
    for (x, y,w, h) in face_rectangle:
        cv2.rectangle(face_img, (x,y), (x+w, y+h), (255, 0,0),4)
    return face_img
```

```python
result = detect_face(solvay)
```

```python
plt.imshow(result, cmap='gray')
```

## Eye Detection

```python
eye_detection = cv2.CascadeClassifier(r'C:\Users\Data\computer vision\haarcascades\haarcascade_eye.xml')
```

```python
def detect_eye(img):
    face_img = img.copy()
    eye_rectangle = eye_detection.detectMultiScale(face_img)
    
    for (x,y,w,h) in eye_rectangle:
        cv2.rectangle(face_img, (x,y),(x+w, y+h), (255, 0,0), 5)
    return face_img
```

```python
result = detect_eye(nadia)
```

```python
plt.imshow(result, cmap='gray')
```

```python
result = detect_eye(denis)

plt.imshow(result, cmap='gray')
```

## Face detection live stream

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    frame = detect_face(frame)
    
    # display
    cv2.imshow('frame',frame)
    
    k = cv2.waitKey(1)
    if k == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
