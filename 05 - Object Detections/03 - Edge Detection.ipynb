```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r'C:\Users\Data\computer vision\sammy_face.jpg')
plt.imshow(img)
```

**edge = cv2.Canny()**

```python
edges = cv2.Canny(img, threshold1 = 127, threshold2= 127)
plt.imshow(edges)
```

we cna see main edges abut also lots of noise, To remove the noise we can :

- blur the image first
- use different threshold value

```python
edges = cv2.Canny(img, threshold1 = 0, threshold2= 255 )
plt.imshow(edges)
```

```python
med_val = np.median(img)
med_val
```

```python
# lower and upper threshold

lower = int(max(0, 0.07 * med_val)) # lower threshold = 0 or 70% of median value
upper = int(min(255, 1.30 * med_val)) # upper threshold = 130 % of median value

edges = cv2.Canny(img, threshold1 = lower, threshold2 = upper)
plt.imshow(edges)
```

```python
edges = cv2.Canny(img, threshold1 = lower, threshold2= upper+75)
plt.imshow(edges)
```

```python
edges = cv2.Canny(img, threshold1 = lower , threshold2= upper +150)
plt.imshow(edges)
```

```python
blured_img = cv2.blur(img, (5,5))

edges = cv2.Canny(blured_img, lower, upper)
plt.imshow(edges)
```

```python
blured_img = cv2.blur(img, ksize=(7,7))

edges = cv2.Canny(blured_img, lower, upper)
plt.imshow(edges)
```
